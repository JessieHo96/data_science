```python
import numpy as np
from matplotlib import pyplot as plt
```

## 绘图基础

```python
#data
x = np.linspace(1,20,100)
y = x ** 2 - 5*x -20
y_ = x**2 - 10*x + 40


figure = plt.figure(figsize=(10,10))

#get current axis
ax = plt.gca()

#spine为坐标轴的框架，可以设定上下左右边框的颜色
ax.spines["right"].set_color("None")
ax.spines['top'].set_color('None')


#设置原点的位置
ax.spines['bottom'].set_position(('data', 100))
ax.spines['left'].set_position(('data', 10))


#设置x,y轴刻度
x_ticks = list(range(1,20,1))
x_ticks = list(range(1,20,1))
y_ticks = np.linspace(y.min(), y.max(), 20)
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)

#画线
plt.plot(x, y, 'r--')
plt.plot(x, y_, 'bv')


#设置x,y轴的标签
plt.xlabel('dependent', loc = 'right', labelpad = 6)
plt.ylabel('independent', loc = 'top', labelpad = 6)


#设置图例
plt.legend(['y1', 'y2'], loc = 'center right')


#设置标注
plt.annotate(r'$this\ is\ an\ annotation$', xy = (x[0], y[0]+20))
```

## 子图

```python
#绘制多个子图方法一：
figure = plt.figure()
ax1 = figure.add_subplot(2,2,1)
ax2 = figure.add_subplot(2,2,2)
ax3 = figure.add_subplot(2,2,3)


ax3.plot(1/2* np.linspace(1,10,10)** 2 - 20)
_ = ax1.hist(np.random.randn(50).cumsum())
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
#设置多个子图方法2
fig, ax = plt.subplots(2,2)
for i in range(2):
    for j in range(2):
        ax[i, j].hist(np.random.randn(500), bins=50, color='r', alpha=0.5)
```

## 常用的图表

```python
#条形图
fig = plt.figure(figsize=(15,5))
fig.add_subplot(2,2,1)

x = np.array(range(1,10))
y = x** + 2*x + 20
y_1 = y + np.random.randint(50,100)

width = 5

plt.plot(x, y, 'kv--')
plt.xticks(x)
plt.bar(x, y)
plt.title("bar")
plt.xlabel("month")
plt.ylabel('incomes')

#水平条形图
fig.add_subplot(2,2,2)

x_tick = np.linspace(y.min(), y.max(), 10)
x = np.array(range(1,10))
y = x** + 2*x + 20

plt.xticks(x_tick)
plt.barh(x, y)
plt.title("hbar")
plt.xlabel("incomes")
plt.ylabel('month')


#展示多个条形图
fig.add_subplot(2,2,3)

width = 0.35

# plt.plot(x, y, 'kv--')
plt.xticks(x)
plt.bar(x, y, width = width, label = 'y')
plt.bar(x + width, y_1, width = width, label = 'y1')
plt.title("bar")
plt.xlabel("month")
plt.ylabel('incomes')

plt.legend(loc = 'upper left')


# 描绘堆叠图
fig.add_subplot(2,2,4)
plt.bar(x, y, label = 'y')
plt.bar(x, y_1, label = 'y1', bottom = y)

plt.legend()
```

```python
#散点图
x = np.linspace(1, 20, 100)

plt.scatter(x, np.sin(x), c = 'r')
plt.scatter(x, np.cos(x), c = 'b')
```

```python
#直方图

figure = plt.figure(figsize=(15, 5))

figure.add_subplot(1,2,1)
x = np.random.randn(500)
plt.hist(x, bins = 10, density = True, rwidth=3)

figure.add_subplot(1,2,2)
x1 = np.random.normal(1, 4, 1000)
x2 = np.random.normal(-1, 2, 1000)
x3 = np.random.normal(0.8, 5, 1000)

pro = dict(alpha=0.4, density=True, bins=20)
plt.hist(x1, **pro)
plt.hist(x2, **pro)
plt.hist(x3, **pro)
plt.legend(['x1', 'x2', 'x3'])
```

```python
#描绘误差条图
x = np.linspace(1,10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y , dy, fmt = '.k')
```

```python
# 描绘堆积图
x = np.array(range(1,10))
y = x** + 2*x + 20
y_1 = y + np.random.randint(50,100)

plt.stackplot(x, y, y_1)

plt.legend(loc = 'upper left', labels = ['y', 'y1'])
```

```python
#描绘热力图
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y= np.random.multivariate_normal(mean, cov, 1000).T
plt.hist2d(x,y, bins=20);
```
